```python
from vllm import LLM, SamplingParams
import os
os.environ["VLLM_LOGGING_LEVEL"] = "DEBUG"  # Включаем подробные логи
os.environ["VLLM_TRACE_FUNCTION"] = "1"     # Включаем трассировку функций (может быть очень шумно)

# ... остальной код ...
# Загружаем модель (один раз, при инициализации)
llm = LLM(model="./LORA_THE_BEST", trust_remote_code=True)

# Параметры генерации
sampling_params = SamplingParams(
    temperature=0.0,
    max_tokens=1600,
    top_p=0.95,
    stop_token_ids=[128001]  # eos_token_id для Llama 3.1
)
```

```python
# Список промптов
prompts = [
    "Мой кариотип 46,XY,der(21;21)(q10;q10),+21. Что это значит?"
]

# Запускаем генерацию
outputs = llm.generate(prompts, sampling_params)

# Выводим результат
for output in outputs:
    print(f"Prompt: {output.prompt}")
    print(f"Generated: {output.outputs[0].text}")
    print("-" * 50)
```
