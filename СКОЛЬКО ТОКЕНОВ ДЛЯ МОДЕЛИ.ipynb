```python
from datasets import load_dataset
from transformers import AutoTokenizer
import numpy as np
import pandas as pd

VAL_PATH = "ПРИМЕРЫ/данные_чистые.xlsx"
MODEL_NAME = "meta-llama/Llama-3.1-8B-Instruct"

# Загружаем токенизатор (без загрузки модели, чтобы сэкономить память)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME, use_fast=True)

# Загружаем датасет
dataset = pd.read_excel(VAL_PATH)
assistant_texts = [k for k in dataset["Ответ"]]

# Подсчитываем количество токенов в каждом ответе
token_counts = [len(tokenizer.encode(text)) for text in assistant_texts]

# Статистика
mean_len = np.mean(token_counts)
median_len = np.median(token_counts)
p95 = np.percentile(token_counts, 95)
p99 = np.percentile(token_counts, 99)
max_len = np.max(token_counts)

print(f"Всего примеров: {len(token_counts)}")
print(f"Среднее: {mean_len:.1f}")
print(f"Медиана: {median_len:.1f}")
print(f"95-й процентиль: {p95:.1f}")
print(f"99-й процентиль: {p99:.1f}")
print(f"Максимум: {max_len}")
```
